```python
import pandas as pd
```

```python
df=pd.read_csv('forecast_inputs.csv', delimiter=';').fillna(0)
df.rename(columns={'Date (WET)': 'Date1'},inplace=True)
df.drop([0],axis=0,inplace=True)
change_date = df.iloc[:,0]
dat=[]
for i in range(1, (len(change_date))):
  value=change_date[i]
  var=value[1:-1]
  dat.append(var)
df['Date'] = pd.to_datetime(pd.Series(dat))
df.drop(['Date1'],axis=1,inplace=True)
df.set_index('Date',inplace=True)
df
```
